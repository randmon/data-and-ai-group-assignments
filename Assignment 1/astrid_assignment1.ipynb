## student info

student gender distribution:
- pretty equal amount of male and female students, with slightly more male students

```python
# look at student gender distribution
print('Number of male students: ', studentinfo_data[studentinfo_data['gender'] == 'M'].shape[0])
print('Number of female students: ', studentinfo_data[studentinfo_data['gender'] == 'F'].shape[0])
sns.catplot(x="gender", kind="count", data=studentinfo_data, palette="ch:.25")
```

student disability distribution
- majority have no disability 

```python
# look at student disability distribution
print('Number of disabled students: ', studentinfo_data[studentinfo_data['disability'] == 'Y'].shape[0])
print('Number of non-disabled students: ', studentinfo_data[studentinfo_data['disability'] == 'N'].shape[0])
sns.catplot(x="disability", kind="count", data=studentinfo_data, palette="ch:.25")
```

look at student age distribution
- most students are ages 0-35
- there are some between 35-55
- small number of students older than 55

```python
# look at student age distribution
print(studentinfo_data['age_band'].value_counts())
sns.catplot(x="age_band", kind="count", data=studentinfo_data, palette="ch:.25")
```

look at student region distribution
- most students are from scotland 
- least students are from ireland

```python
# look at student location distribution sort by number of students
# print student region by count
print(studentinfo_data['region'].value_counts())
sns.catplot(x="region", kind="count", data=studentinfo_data, palette="ch:.25", aspect=3, order=studentinfo_data['region'].value_counts().index)
```

look at student highest education
- most have a level, next is those who have lower
- large amount of he qualification
- small amount of people with post graduate qualification or no formal qualification

```python
# look at student highest education distribution
sns.catplot(x="highest_education", kind="count", data=studentinfo_data, palette="ch:.25", aspect=3)
```

look at imd_band distribution

```python
# look at student imd_band distribution
print(studentinfo_data['imd_band'].value_counts())

sns.catplot(x="imd_band", kind="count", data=studentinfo_data, palette="ch:.25", aspect=3)
```

look at students number of previous attempts distribution
- most students have 0 previous attempts
- as amount of attempts increases, amount of students decreases

```python
# look at students number of previous attempts distribution
print(studentinfo_data['num_of_prev_attempts'].value_counts())
sns.catplot(x="num_of_prev_attempts", kind="count", data=studentinfo_data, palette="ch:.25")
```

look at students final result distribution
- most students pass

```python
# look at students final result distribution
print(studentinfo_data['final_result'].value_counts())
sns.catplot(x="final_result", kind="count", data=studentinfo_data, palette="ch:.25")
```

#### 3) explore the data:

```python
import pandas as pd # Common convention

# we import the information about the students. 
students_df = pd.read_csv('studentInfo.csv')
students_df = students_df[students_df.columns[students_df.columns.isin(["id_student", "region", "highest_education","num_of_prev_attempts", "studied_credits", "final_result"])]]

# we import the information that describes, the students interaction with the materials. 
studentsI_df = pd.read_csv('studentVle.csv')
studentsI_df = studentsI_df[studentsI_df.columns[studentsI_df.columns.isin(["id_student","sum_click"])]]

studentAssessment_df = pd.read_csv('studentAssessment.csv')
studentAssessment_df = studentAssessment_df[studentAssessment_df.columns[studentAssessment_df.columns.isin(["id_assessment","score"])]]

assessments_df = pd.read_csv('assessments.csv')
assessments_df = assessments_df[assessments_df.columns[assessments_df.columns.isin(["id_assessment","assessment_type"])]]
```

```python
# lets also merge these tables 

merged_df = pd.merge(studentsI_df, students_df, on=["id_student"])
merged_df
```

```python
# people who interact with the course more seem to perform better

merged_df.groupby('final_result', as_index=False)['sum_click'].mean()
```

```python
# people with lover credit seem to perform better
# the people who join with a lot of credits keep failing!
merged_df.groupby('final_result', as_index=False)['studied_credits'].mean()
```

```python
# the highest achiever usually pas from the first time
merged_df.groupby('final_result', as_index=False)['num_of_prev_attempts'].mean()
```

```python
merged_assessment_df = pd.merge(studentAssessment_df, assessments_df, on=["id_assessment"])
```

```python
# people seem to do the worst at exams
merged_assessment_df.groupby('assessment_type', as_index=False)['score'].mean()
```
